```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Đọc dữ liệu từ file excel
data = pd.read_csv('goldData.csv',index_col=0)
```

```python
data.count()
```

```python
spreadProfile = data['spreadProfile'].value_counts()
spreadProfile
```

```python
plt.bar(spreadProfile.index,spreadProfile)
plt.title('spreadProfile')
```

```python
platform = data['platform'].value_counts()
plt.bar(platform.index,platform)
plt.title("Platform")
```

```python
server = data['server'].value_counts()
plt.bar(server.index,server)
plt.title("Server")
```

```python
bidSpread = data['bidSpread']
askSpread = data['askSpread']
bid = data['bid']
ask = data['ask']
```

```python
var = pd.DataFrame(data.groupby(['server'])['bidSpread', 'askSpread'].mean().sort_values(by= "bidSpread", ascending=False))
var.plot.bar()
var = pd.DataFrame(data.groupby(['server'])['bid', 'ask'].mean().sort_values(by= "bid", ascending=False))
var.plot.line()
```

```python
var = pd.DataFrame(data.groupby(['platform'])['bidSpread', 'askSpread'].mean().sort_values(by= "bidSpread", ascending=False))
var.plot.line()
var = pd.DataFrame(data.groupby(['platform'])['bid', 'ask'].mean().sort_values(by= "bid", ascending=False))
var.plot.line()
```

```python
var = pd.DataFrame(data.groupby(['spreadProfile'])['bidSpread', 'askSpread'].mean().sort_values(by= "bidSpread", ascending=False))
var.plot.bar()
var = pd.DataFrame(data.groupby(['spreadProfile'])['bid', 'ask'].mean().sort_values(by= "bid", ascending=False))
var.plot.line()
```
